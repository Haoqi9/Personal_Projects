# Business Problem

`Contexto hipotético`:

Hypothetical Context:

- A financial institution wants to **improve its risk assessment process and reduce losses associated with payment default**. To address this challenge, it decides to develop a credit default detection algorithm that can accurately identify customers who are most likely to default on their financial obligations.

- **The objective is to prevent granting credits to individuals with a high risk of default**, while maximizing business opportunities by granting credits to those with a low risk of default. *We will decide later whether it is more convenient to develop a credit scoring model focused on new customers or on existing customers of the institution*.

In this case, **a dataset of loans granted** by a US financial institution will be used. This dataset comes from an online course on data science: *Data Science For Business*.

# Define Target Variable <a id=target><a />

```python
df['estado'].value_counts(dropna=False, normalize=True)
```

```python
class_neg = ["Fully Paid", "Current", "In Grace Period", "Late (16-30 days)", "Does not meet the credit policy. Status:Fully Paid"]
class_pos = ["Charged Off", "Late (31-120 days)", "Does not meet the credit policy. Status:Charged Off", "Default"]

df['default'] = np.where(df['estado'].isin(class_pos), 1, 0)
df['default'].value_counts(dropna=False, normalize=True)
```

Observations:
- Given the categorical variable `estado` in the dataset, a binary variable `default` has been created to represent whether a borrowing customer has defaulted on a credit or not:
  
    - Since we do not have specific information in the company regarding how they define what a credit default is, **I have made my own decisions about it: those with a payment delay of more than 30 days after the agreed deadline**. Initially, I was considering choosing a threshold of 60-90 days, but since the category is overlapped in the "Late (31-120 days)" category, *I decided to cut at 30 days since the goal is to minimize credit risk as much as possible, and a payment delay of up to 120 days is an indicator of high risk of default*.
      
    - The **negative class (0)** has been assigned to the categories "Fully Paid", "Current", "In Grace Period", "Late (16-30 days)", "Does not meet the credit policy. Status:Fully Paid".
      
    - The **positive class (1)** has been assigned to: "Charged Off", "Late (31-120 days)", "Does not meet the credit policy. Status:Charged Off", "Default".

- **We are dealing with a binary classification problem with class imbalance** since only about 13% of borrowers have defaulted (according to my definition) compared to the 87% who did not.

## imp_amortizado

```python
df.groupby('default')['imp_amortizado'].median()
```

```python
# TODO: eliminar imp_recuperado e imp_amortizado:
df.drop(columns=["imp_recuperado", "imp_amortizado"], inplace=True)
df.columns
```

- We can infer that **these two variables contain information after the client has paid off the entire loan or has been considered "Default" or "Charged Off"**. This information will not be available at prediction time. Additionally, it can be observed that on median, nothing has been recovered from the loan for clients who have complied with the payment, while for those who default, a median of $452 has been recovered! Including it as a predictor would be an issue (information leakage).

- Indeed, **such variable shouldn't be used along with others like "dti," "rating," "imp_cuota," "imp_amortizado," and others**. The reason is that *if we have a new client, we won't have a credit history (these variables) about them. Therefore, they cannot be used for prediction*.

- Taking this into account, **not using these variables specific to veteran clients may result in an inadequate credit scoring model**. *It may even be of interest to us that it is a scoring model only for old clients*. If, however, these variables regarding the client's credit history turn out to be uninformative, then focusing on new clients might be more appropriate. `We will decide in the evaluation phase!`

## Group infrequent categories <a id=infrequent><a />

- We're going to **group infrequent categories after null imputation because some categorical variables had null values**. If we had grouped infrequent categories in those variables with nulls by mode, it could happen that the category, let's say "other," to which many infrequent categories have been joined becomes the most frequent one, **biasing that imputation**.

```python
pd.concat(
    [
        pd.Series(df_train.rating.value_counts(normalize=False), name='count'),
        pd.Series(df_train.rating.value_counts(normalize=True).round(2), name='freq')
    ],
    axis=1
)
```

## One-Hot Encoding

```python
# TODO: ohe.
print(f"Numero de columnas antes: {df_train.shape[1]}.")

ohe_vars = ["antigüedad_empleo", "ingresos_verificados", "rating", "vivienda", "finalidad", "num_cuotas"]
df_train = pd.get_dummies(
    data=df_train,
    columns=ohe_vars,
    drop_first=False,
    dtype=float,
)
print(f"Numero de columnas después: {df_train.shape[1]}.")
```

## TF-IDF Vectorization

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
import re

# TODO: Convertir texto en tokens normalizados:
stopwords_en  = stopwords.words('english')
SStemmer = SnowballStemmer(language='english')

def preprocess_text(sentence):
    # Eliminamos dígitos y puntuaciones.
    text   = re.sub(pattern='[\d\W]',repl=' ', string=sentence)
    text   = text.lower()
    tokens = word_tokenize(text)
    # Eliminar stopwords y lematizar.
    tokens = [SStemmer.stem(token) for token in tokens if token not in stopwords_en]
    # Eliminar extra spaces.
    tokens = [token.strip() for token in tokens]
    # Eliminar '' token.
    tokens = [token for token in tokens if token != '']
    return ' '.join(tokens)

df_train['empleo_token'] = df_train.empleo.apply(lambda text: preprocess_text(text))
```

```python
df_train[['empleo', 'empleo_token']].sample(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TODO: vectorizar cada token.
vectorizador_tfidf = TfidfVectorizer(
    lowercase=False,
    stop_words =None,
    min_df=0.005,
    ngram_range = (1,3),
    use_idf=True,
    smooth_idf=True,
)

empleo_vectorized = vectorizador_tfidf.fit_transform(df_train['empleo_token'])
empleo_features   = pd.DataFrame(empleo_vectorized.toarray(), columns=vectorizador_tfidf.get_feature_names_out())
empleo_features.shape
```
